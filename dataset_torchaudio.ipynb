```python
train_list = []
test_list = []
validation_list = []

for artist in artist_list:
    path_album = os.path.join(song_path, artist)
    songs_by_artist = os.listdir(path_album)
    song_path_by_artist = [os.path.join(path_album, song) for song in songs_by_artist]
    song_path_by_artist_array = np.array(song_path_by_artist)
    np.random.shuffle(song_path_by_artist_array)
    X_train, X_test = train_test_split(song_path_by_artist_array, train_size=0.90, shuffle=True, random_state=43)
    X_train, X_valition = train_test_split(X_train, train_size=0.90, shuffle=True, random_state=43)
    train_list.append((X_train, artist))
    test_list.append((X_test, artist))
    validation_list.append((X_valition, artist))

    
```

```python
train_array = [{"song_path": song, "artist": artist} for songs, artist in train_list for song in songs]
test_array = [{"song_path": song, "artist": artist} for songs, artist in test_list for song in songs]
validation_array = [{"song_path": song, "artist": artist} for songs, artist in validation_list for song in songs]
```

```python
train = pd.DataFrame(train_array)
test = pd.DataFrame(test_array)
validation = pd.DataFrame(validation_array)
```

```python
def plot_spectrogram(specgram, title=None, ylabel="freq_bin", ax=None):
    if ax is None:
        _, ax = plt.subplots(1, 1)
    if title is not None:
        ax.set_title(title)
    ax.set_ylabel(ylabel)
    ax.imshow(librosa.power_to_db(specgram), origin="lower", aspect="auto", interpolation="nearest")
    # ax.colorbar(format='%+2.0f dB')


def plot_waveform(waveform, sr, title="Waveform", ax=None):
    waveform = waveform.numpy()

    num_channels, num_frames = waveform.shape
    time_axis = torch.arange(0, num_frames) / sr

    if ax is None:
        _, ax = plt.subplots(num_channels, 1)
    ax.plot(time_axis, waveform[0], linewidth=1)
    ax.grid(True)
    ax.set_xlim([0, time_axis[-1]])
    ax.set_title(title)
```

Parallel Process 

```python
import pandas as pd
import concurrent.futures
import os

time_frame = [5]
# Specify the directory path
directory_path = './data'

# List files in the directory
files = os.listdir(directory_path)

# Define the number of worker threads
num_workers = len(files)

# Create a ThreadPoolExecutor
with concurrent.futures.ThreadPoolExecutor(max_workers = num_workers) as executor:
    # Submit tasks to read CSV files in parallel
    futures = {executor.submit(execute_algorithm_create_datasets ,directory_path, file_name, time_frame): file_name for file_name in files}

    # # Retrieve the results as they become available
    for future in concurrent.futures.as_completed(futures):
        file_name = futures[future]
        response =  future.result()
        print(response)
```

Collect data from folders after spliting

```python

import os

def spectrogram_to_csv_by_file(path_to_read:str = "", path_to_save:str = "") -> None:
    import os
    from tqdm import tqdm
    import gzip
    import pickle
    import pandas as pd
    try:
        if len(path_to_read)!=0 or len(path_to_save)!=0:
            
            listFiles = os.listdir(path_to_read)
            # print(path_to_read, path_to_save, "mierda", len(listFiles))
            pbar = tqdm(listFiles)
            data = []
            for i in pbar:
                file_path = os.path.join(path_to_read, i)
                file_path = file_path.replace('\\', '/')
                with gzip.open(file_path, 'rb') as f:
                    data_file = pickle.load(f)
    
                data.append({
                    'label': data_file['artist_name'],
                    'path': file_path
                    })
            print("Converting data to csv")
            dataFrame = pd.DataFrame(data)
            dataFrame.to_csv(path_to_save)
            print(f"Done file {path_to_save} save")
    except:
        raise("Error please enter data path_to_read and path_to_save")
```

```python
datasets = ['train', 'test', 'validation']
dataPath = "./dataset"

for i in datasets:
    print(f"Working on {i} dataset.")
    path_to_read = os.path.join(dataPath, i)
    path_to_read = path_to_read.replace('\\', '/')

    path_to_save = os.path.join(dataPath, f"{i}.csv")
    path_to_save = path_to_save.replace('\\', '/')

    
    spectrogram_to_csv_by_file(path_to_read = path_to_read, path_to_save = path_to_save)
    print(f"{i} finished.")
    
```
